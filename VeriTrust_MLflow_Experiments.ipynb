```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from tqdm import tqdm

CLASS_NAMES  = ["REAL", "AI_GEN", "PHOTOSHOP"]
CLASS_COLORS = {"REAL": "#2196F3", "AI_GEN": "#FF5722", "PHOTOSHOP": "#4CAF50"}

DATA_DIR = Path("data/processed")
folder_map = {
    "REAL"      : DATA_DIR / "real",
    "AI_GEN"    : DATA_DIR / "ai_generated",
    "PHOTOSHOP" : DATA_DIR / "photoshop",
}


def load_all_features(folder_map: dict) -> tuple:
    """Return (X [NxNUM_FEATURES], labels [N]) for all images in folder_map."""
    all_feats, all_labels = [], []
    for cls_name, folder in folder_map.items():
        imgs = list(folder.glob("*.jpg")) + list(folder.glob("*.png"))
        for path in tqdm(imgs, desc=f"  {cls_name:<12}"):
            try:
                img   = Image.open(path).convert("RGB").resize((256, 256))
                feats = extract_features(img)
                all_feats.append(feats)
                all_labels.append(cls_name)
            except Exception as e:
                print(f"  skip {path.name}: {e}")
    return np.array(all_feats), all_labels


if __name__ == "__main__":
    print("\nExtracting features from all images...")
    X, labels = load_all_features(folder_map)
    print(f"Total : {len(X)} images  |  shape : {X.shape}")
```

```python
# ── Gather all image paths and labels ────────────────────────────────────
from sklearn.model_selection import train_test_split

SEED = 42

all_paths  = []
all_lbls   = []
for class_id, folder in folder_map.items():
    for p in sorted(folder.glob("*.jpg")) + sorted(list(folder.glob("*.png"))):
        all_paths.append(str(p))
        all_lbls.append(class_id)

all_paths = np.array(all_paths)
all_lbls  = np.array(all_lbls)

# ── Train / val / test split: 70 / 15 / 15 ───────────────────────────────
idx = np.arange(len(all_paths))
idx_tv, idx_test = train_test_split(idx, test_size=0.15, stratify=all_lbls, random_state=SEED)
idx_train, idx_val = train_test_split(idx_tv, test_size=0.176, stratify=all_lbls[idx_tv], random_state=SEED)

print(f"Train : {len(idx_train)} images")
print(f"Val   : {len(idx_val)}   images")
print(f"Test  : {len(idx_test)}  images")
```

```python
# ── Imports ──────────────────────────────────────────────────────────────
import numpy as np
import platform

import torch
import torchvision.transforms as T

from torch.utils.data import (
    Dataset,
    DataLoader,
    WeightedRandomSampler
)

from sklearn.preprocessing import LabelEncoder
from PIL import Image

# On Windows, num_workers > 0 crashes in notebooks
NUM_WORKERS = 0 if platform.system() == "Windows" else 2

# ── Constants ────────────────────────────────────────────────────────────
IMG_SIZE = 224

BATCH_SIZE = 16

# ============================================================
# FIX LABELS
# ============================================================
# all_lbls contains STRING labels
# Example:
# ["Real", "Fake", "AI"]

label_encoder = LabelEncoder()

# Convert labels → integers
encoded_labels = label_encoder.fit_transform(all_lbls)

# Total classes
NUM_CLASSES = len(label_encoder.classes_)

print("Encoded Classes:")

for idx, class_name in enumerate(label_encoder.classes_):

    print(f"{class_name} -> {idx}")

# ── Image transforms ─────────────────────────────────────────────────────
train_transform = T.Compose([

    T.Resize((IMG_SIZE, IMG_SIZE)),

    T.RandomHorizontalFlip(),

    T.ColorJitter(
        brightness=0.15,
        contrast=0.15,
        saturation=0.10
    ),

    T.RandomRotation(8),

    T.ToTensor(),

    T.Normalize(
        [0.485, 0.456, 0.406],
        [0.229, 0.224, 0.225]
    ),
])

val_transform = T.Compose([

    T.Resize((IMG_SIZE, IMG_SIZE)),

    T.ToTensor(),

    T.Normalize(
        [0.485, 0.456, 0.406],
        [0.229, 0.224, 0.225]
    ),
])

# ── Dataset class ────────────────────────────────────────────────────────
class VeriTrustDataset(Dataset):
    """
    Returns:
        image_tensor,
        feature_vector,
        label
    """

    def __init__(
        self,
        paths,
        labels,
        transform
    ):

        self.paths = paths

        self.labels = labels

        self.transform = transform

    def __len__(self):

        return len(self.paths)

    def __getitem__(self, idx):

        # ── Load image ──────────────────────────────
        img = Image.open(
            self.paths[idx]
        ).convert("RGB")

        # ── Hand-crafted features ───────────────────
        feats = extract_features(

            img.resize((256, 256))

        )

        # Convert features → tensor
        feats = torch.tensor(
            feats,
            dtype=torch.float32
        )

        # ── CNN image transform ─────────────────────
        img_t = self.transform(img)

        # Integer label
        label = int(
            self.labels[idx]
        )

        return img_t, feats, label

# ============================================================
# BUILD DATASETS
# ============================================================

train_ds = VeriTrustDataset(

    all_paths[idx_train],

    encoded_labels[idx_train],

    train_transform
)

val_ds = VeriTrustDataset(

    all_paths[idx_val],

    encoded_labels[idx_val],

    val_transform
)

test_ds = VeriTrustDataset(

    all_paths[idx_test],

    encoded_labels[idx_test],

    val_transform
)

# ============================================================
# WEIGHTED SAMPLER (CLASS BALANCING)
# ============================================================

train_lbls = encoded_labels[idx_train]

# Count class frequencies
class_cnts = np.bincount(

    train_lbls,

    minlength=NUM_CLASSES
)

print("\nClass counts:")

print(class_cnts)

# Avoid division by zero
class_wts = 1.0 / np.maximum(class_cnts, 1)

# Assign weight to each sample
sample_wts = class_wts[
    train_lbls
]

# Convert → tensor
sample_wts = torch.DoubleTensor(
    sample_wts
)

sampler = WeightedRandomSampler(

    weights=sample_wts,

    num_samples=len(sample_wts),

    replacement=True
)

# ============================================================
# DATALOADERS
# ============================================================

train_loader = DataLoader(

    train_ds,

    batch_size=BATCH_SIZE,

    sampler=sampler,

    num_workers=NUM_WORKERS,

    pin_memory=(NUM_WORKERS > 0),
)

val_loader = DataLoader(

    val_ds,

    batch_size=BATCH_SIZE,

    shuffle=False,

    num_workers=NUM_WORKERS,

    pin_memory=(NUM_WORKERS > 0),
)

test_loader = DataLoader(

    test_ds,

    batch_size=BATCH_SIZE,

    shuffle=False,

    num_workers=NUM_WORKERS,

    pin_memory=(NUM_WORKERS > 0),
)

# ============================================================
# SUMMARY
# ============================================================

print(
    f"\nLoaders ready — "
    f"{len(train_loader)} train batches, "
    f"{len(val_loader)} val batches, "
    f"{len(test_loader)} test batches"
)
```

```python
# ── 13 Hand-crafted detection features ──────────────────────────────────
# Removed: brightness_mean, brightness_std, contrast_rms  (weak, overlapping)
# Added  : fft_energy_ratio, laplacian_variance, lbp_entropy, residual_entropy

from skimage.feature import local_binary_pattern  # pip install scikit-image

def extract_features(pil_img: Image.Image) -> np.ndarray:
    """
    Extract 13 statistical features from a PIL image.

    Features
    --------
    0  noise_sigma        — estimated sensor noise (high-freq Gaussian residual)
    1  edge_density       — Laplacian variance / 1e6  (splice hard-edge detector)
    2  fft_energy_ratio   — ratio of high-freq energy to total energy in FFT
                            (AI images have unnaturally smooth spectra)
    3  laplacian_variance — raw Laplacian variance (sharpness; complements edge_density)
    4  lbp_entropy        — Shannon entropy of LBP histogram
                            (texture regularity; AI textures are too uniform)
    5  jpeg_block_score   — mean high-freq DCT energy in 8×8 blocks
                            (JPEG re-compression artifact strength)
    6  residual_entropy   — entropy of the high-freq noise residual
                            (AI noise is structured; real camera noise is random)
    7  r_mean             — mean Red channel
    8  g_mean             — mean Green channel
    9  b_mean             — mean Blue channel
    10 colour_entropy     — Shannon entropy of quantised luminance histogram
    11 local_var_cv       — CoV of 16×16 patch variances
                            (low = copy-paste regions from Photoshop)
    12 r_g_diff           — |R_mean − G_mean| colour cast indicator
    """
    img_rgb  = np.array(pil_img.convert("RGB"), dtype=np.float32)
    img_gray = np.array(pil_img.convert("L"),   dtype=np.float32)
    h, w     = img_gray.shape

    # ── 0: noise sigma ────────────────────────────────────────────────────
    blurred     = cv2.GaussianBlur(img_gray, (5, 5), 0)
    residual    = img_gray - blurred
    noise_sigma = residual.std() / 255.0

    # ── 1: edge density (Laplacian variance, normalised) ──────────────────
    lap          = cv2.Laplacian(img_gray.astype(np.uint8), cv2.CV_64F)
    edge_density = lap.var() / 1e6

    # ── 2: FFT energy ratio (high-freq / total) ───────────────────────────
    fft         = np.fft.fft2(img_gray)
    fft_shifted = np.fft.fftshift(fft)
    magnitude   = np.abs(fft_shifted) ** 2
    cy, cx      = h // 2, w // 2
    radius      = min(h, w) // 4          # inner 25% = low freq
    Y, X        = np.ogrid[:h, :w]
    dist_mask   = (X - cx) ** 2 + (Y - cy) ** 2
    low_freq_mask  = dist_mask <= radius ** 2
    total_energy   = magnitude.sum() + 1e-9
    high_freq_energy = magnitude[~low_freq_mask].sum()
    fft_energy_ratio = float(high_freq_energy / total_energy)

    # ── 3: laplacian variance (raw, not / 1e6) ────────────────────────────
    laplacian_variance = float(lap.var())

    # ── 4: LBP entropy ────────────────────────────────────────────────────
    lbp        = local_binary_pattern(img_gray.astype(np.uint8), P=8, R=1, method="uniform")
    lbp_hist, _= np.histogram(lbp.ravel(), bins=10, range=(0, 10), density=True)
    lbp_hist   = lbp_hist + 1e-9
    lbp_entropy = float(-np.sum(lbp_hist * np.log2(lbp_hist)) / np.log2(10))

    # ── 5: JPEG 8×8 DCT block artifact score ─────────────────────────────
    block_scores = []
    for y in range(0, h - 8 + 1, 8):
        for x in range(0, w - 8 + 1, 8):
            block = img_gray[y:y+8, x:x+8]
            dct   = cv2.dct(block.astype(np.float32))
            block_scores.append(np.abs(dct[4:, 4:]).mean())
    jpeg_block_score = np.mean(block_scores) / 255.0

    # ── 6: residual entropy ───────────────────────────────────────────────
    res_uint8      = np.clip(residual + 128, 0, 255).astype(np.uint8)
    res_hist, _    = np.histogram(res_uint8.ravel(), bins=32, range=(0, 256), density=True)
    res_hist       = res_hist + 1e-9
    residual_entropy = float(-np.sum(res_hist * np.log2(res_hist)) / np.log2(32))

    # ── 7, 8, 9: per-channel means ────────────────────────────────────────
    r_mean = img_rgb[:, :, 0].mean() / 255.0
    g_mean = img_rgb[:, :, 1].mean() / 255.0
    b_mean = img_rgb[:, :, 2].mean() / 255.0

    # ── 10: colour histogram entropy ──────────────────────────────────────
    hist, _ = np.histogram(img_gray.flatten(), bins=32, range=(0, 256))
    hist    = hist / hist.sum() + 1e-9
    colour_entropy = float(-np.sum(hist * np.log2(hist)) / np.log2(32))

    # ── 11: local variance coefficient of variation ───────────────────────
    patch = 16
    variances = []
    for y in range(0, h - patch + 1, patch):
        for x in range(0, w - patch + 1, patch):
            variances.append(img_gray[y:y+patch, x:x+patch].var())
    variances    = np.array(variances) + 1e-9
    local_var_cv = float(variances.std() / variances.mean())

    # ── 12: R–G channel difference ────────────────────────────────────────
    r_g_diff = float(abs(r_mean - g_mean))

    return np.array([
        noise_sigma, edge_density,
        fft_energy_ratio, laplacian_variance,
        lbp_entropy, jpeg_block_score, residual_entropy,
        r_mean, g_mean, b_mean,
        colour_entropy, local_var_cv, r_g_diff,
    ], dtype=np.float32)


FEATURE_NAMES = [
    "noise_sigma", "edge_density",
    "fft_energy_ratio", "laplacian_variance",
    "lbp_entropy", "jpeg_block_score", "residual_entropy",
    "r_mean", "g_mean", "b_mean",
    "colour_entropy", "local_var_cv", "r_g_diff",
]
NUM_FEATURES = len(FEATURE_NAMES)
print(f"Feature extractor ready — {NUM_FEATURES} features per image")
print("\nFeatures:")
for i, name in enumerate(FEATURE_NAMES):
    print(f"  [{i:02d}] {name}")
```

```python
# ============================================================
# TASK B-4 — MODEL VERSIONING & EXPERIMENTATION
# ============================================================

import mlflow
import mlflow.pytorch
import cv2

import torch
import torch.nn as nn
import torch.optim as optim

import torchvision.models as models

from sklearn.metrics import accuracy_score

import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# DEVICE
# ============================================================

DEVICE = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print("Using device:", DEVICE)

# ============================================================
# VERIFY EXISTING DATALOADERS
# ============================================================

print("Train batches:", len(train_loader))
print("Validation batches:", len(val_loader))

# ============================================================
# SETUP MLFLOW
# ============================================================

mlflow.set_experiment(
    "VeriTrust_Experiments"
)

# ============================================================
# MODEL ARCHITECTURE
# IMAGE-ONLY VERSION FOR MLFLOW TESTING
# ============================================================

class VeriTrustModel(nn.Module):

    def __init__(self, dropout_rate=0.3):

        super().__init__()

        self.model = models.mobilenet_v3_small(
            weights="DEFAULT"
        )

        # CORRECT FEATURE SIZE
        in_features = self.model.classifier[0].in_features

        # Replace classifier
        self.model.classifier = nn.Sequential(

            nn.Linear(in_features, 256),

            nn.Hardswish(),

            nn.Dropout(dropout_rate),

            nn.Linear(256, 3)
        )

    def forward(self, x):

        return self.model(x)

# ============================================================
# LOSS FUNCTION
# ============================================================

criterion = nn.CrossEntropyLoss()

# ============================================================
# TRAINING FUNCTION
# ============================================================

def train_model(

    model,
    train_loader,
    val_loader,
    epochs,
    learning_rate
):

    optimizer = optim.Adam(

        model.parameters(),
        lr=learning_rate
    )

    model.to(DEVICE)

    best_accuracy = 0

    for epoch in range(epochs):

        # ====================================================
        # TRAINING
        # ====================================================

        model.train()

        running_loss = 0

        # Dataset returns:
        # image, features, label
        for images, _, labels in train_loader:

            images = images.to(DEVICE)

            labels = labels.to(DEVICE)

            optimizer.zero_grad()

            # IMAGE ONLY
            outputs = model(images)

            loss = criterion(
                outputs,
                labels
            )

            loss.backward()

            optimizer.step()

            running_loss += loss.item()

        # ====================================================
        # VALIDATION
        # ====================================================

        model.eval()

        y_true = []

        y_pred = []

        with torch.no_grad():

            for images, _, labels in val_loader:

                images = images.to(DEVICE)

                labels = labels.to(DEVICE)

                outputs = model(images)

                preds = torch.argmax(
                    outputs,
                    dim=1
                )

                y_true.extend(
                    labels.cpu().numpy()
                )

                y_pred.extend(
                    preds.cpu().numpy()
                )

        accuracy = accuracy_score(
            y_true,
            y_pred
        )

        print(
            f"Epoch {epoch+1}/{epochs} "
            f"| Loss: {running_loss:.4f} "
            f"| Accuracy: {accuracy:.4f}"
        )

        if accuracy > best_accuracy:

            best_accuracy = accuracy

    return best_accuracy, model

# ============================================================
# EXPERIMENT CONFIGURATIONS
# ============================================================

experiment_configs = [

    {
        "run_name": "Model_V1",
        "learning_rate": 1e-3,
        "dropout": 0.3,
        "epochs": 5
    },

    {
        "run_name": "Model_V2",
        "learning_rate": 5e-4,
        "dropout": 0.4,
        "epochs": 5
    },

    {
        "run_name": "Model_V3",
        "learning_rate": 1e-4,
        "dropout": 0.5,
        "epochs": 5
    }
]

# ============================================================
# RUN EXPERIMENTS
# ============================================================

results = []

best_model = None

best_accuracy = 0

for config in experiment_configs:

    with mlflow.start_run(
        run_name=config["run_name"]
    ):

        print("\n===================================")

        print(
            "Starting:",
            config["run_name"]
        )

        print("===================================")

        # ====================================================
        # CREATE MODEL
        # ====================================================

        model = VeriTrustModel(

            dropout_rate=config["dropout"]
        )

        # ====================================================
        # TRAIN MODEL
        # ====================================================

        accuracy, trained_model = train_model(

            model=model,

            train_loader=train_loader,

            val_loader=val_loader,

            epochs=config["epochs"],

            learning_rate=config["learning_rate"]
        )

        # ====================================================
        # LOG PARAMETERS
        # ====================================================

        mlflow.log_param(
            "learning_rate",
            config["learning_rate"]
        )

        mlflow.log_param(
            "dropout",
            config["dropout"]
        )

        mlflow.log_param(
            "epochs",
            config["epochs"]
        )

        # ====================================================
        # LOG METRICS
        # ====================================================

        mlflow.log_metric(
            "validation_accuracy",
            accuracy
        )

        # ====================================================
        # LOG MODEL
        # ====================================================

        mlflow.pytorch.log_model(
            trained_model,
            "model"
        )

        # ====================================================
        # SAVE RESULTS
        # ====================================================

        results.append({

            "Model": config["run_name"],

            "Learning Rate": config["learning_rate"],

            "Dropout": config["dropout"],

            "Epochs": config["epochs"],

            "Validation Accuracy": accuracy
        })

        # ====================================================
        # TRACK BEST MODEL
        # ====================================================

        if accuracy > best_accuracy:

            best_accuracy = accuracy

            best_model = trained_model

        print(
            "Completed:",
            config["run_name"]
        )

# ============================================================
# RESULTS TABLE
# ============================================================

results_df = pd.DataFrame(results)

display(results_df)

# ============================================================
# VISUALIZE RESULTS
# ============================================================

plt.figure(figsize=(8,5))

plt.bar(

    results_df["Model"],

    results_df["Validation Accuracy"]
)

plt.title(
    "Validation Accuracy Comparison"
)

plt.xlabel("Model")

plt.ylabel("Accuracy")

plt.show()

# ============================================================
# SAVE BEST MODEL
# ============================================================

torch.save(

    best_model.state_dict(),

    "best_veritrust_model.pth"
)

print("\nBest model saved successfully.")

# ============================================================
# MLFLOW UI INSTRUCTIONS
# ============================================================

print("\nRun this in terminal:")

print("mlflow ui")

print("\nOpen browser:")

print("http://127.0.0.1:5000")
```
